This is a simple Python notebook. Below we'll show off matplotlib plotting.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.ion()
```

```python
x = 100 * np.random.rand(500)
y = 100 * np.random.rand(500)
areas = 1000 * np.random.rand(500)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(x, y, s=areas, c=areas, alpha=.5)
```
